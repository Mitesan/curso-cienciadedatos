```python
import pandas as pd
pd.options.display.max_columns = None #Para que se visualicen todas las columnas del dataframe
df = pd.read_csv('cars.csv')
df
```

# One-Hot 

```python
pd.get_dummies(df['engine_type']) #procesado one-hot con Pandas
```

```python
import sklearn.preprocessing as preprocessing
encoder = preprocessing.OneHotEncoder(handle_unknown='ignore') #procesado one-hot con Sklearn
```

```python
encoder.fit(df[['engine_type']].values) # ajustar el encoder a las categorias del dataset
```

```python
encoder.transform([['gasoline'],['diesel'],['agua']]).toarray()
```

Notamos que el tipo de motor "agua" se le asigna el valor 0,0,0 dado que no existe ningún dato con ese valor.
