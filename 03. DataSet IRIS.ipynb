```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("seaborn-dark")
```

```python
data = pd.read_csv('iris.csv', index_col = 0)
data.head()
```

loc $ iloc
- loc utiliza la etiqueta
- iloc utiliza la posición

```python
plt.hist(data.iloc[:,0], edgecolor='black')
plt.title('Distribución del largo del Sépalo en cm')
plt.show()
```

```python
#La media (promedio) del largo del sépalo
data.iloc[:,0].mean()
#data['SepalLengthCm'].mean()
```

```python
plt.hist(data.iloc[:,1], edgecolor='black')
plt.title('Distribución del ancho del Sépalo en cm')
plt.show()
```

```python
#La media (promedio) del ancho del sépalo
data.iloc[:,1].mean()
```

```python
plt.hist(data.iloc[:,2], edgecolor='black')
plt.title('Distribución del largo del Pétalo en cm')
plt.show()
```

```python
#La media (promedio) del largo del pétalo
data.iloc[:,2].mean()
```

```python
plt.hist(data.iloc[:,3], edgecolor='black')
plt.title('Distribución del ancho del Pétalo en cm')
plt.show()
```

```python
#La media (promedio) del ancho del pétalo
data.iloc[:,3].mean()
```

```python
#Distribuciones
plt.scatter(data['PetalLengthCm'], data['PetalWidthCm']) #X #Y
plt.title('DImensiones de los Pétalos')
plt.show()

#comando para hacer una linea de promedio en el eje x
#plt.axvline(df.iloc[:,0].mean(), color='k', linestyle='dashed', linewidth=1)

#para el promedio del eje Y
#plt.axhline(df.iloc[:,0].mean(), color='blue', linestyle='dashed', linewidth=2)
```

```python
indices = data[data['PetalLengthCm']>=2.5].index
indices
```

```python
data.loc[indices,'Tamaño'] = 'Grande'
data.loc[indices, 'Color'] = 'r'
data
```

```python
#Para el caso contrario:
indices = data[data['PetalLengthCm']<2.5].index
data.loc[indices,'Tamaño'] = 'Pequeña'
data.loc[indices, 'Color'] = 'b'
data
```

```python
#Incluyendo el color
plt.scatter(data['PetalLengthCm'], data['PetalWidthCm'], c=data.loc[:,'Color']) #X, #Y
plt.title('DImensiones de los Pétalos')
plt.show()
```
