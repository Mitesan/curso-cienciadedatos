```python
# Importar librerias
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("seaborn-dark")
import seaborn as sns
from scipy import stats

data = pd.read_csv('mpg.csv')
data
```

```python
data.info()
```

```python
# Matriz de Correlaciones
sns.heatmap(data.corr(), annot=True)
```

De la matriz de correlaciÃ³n obervamos que las variables "displacement" y "weight" tienen un coeficiente de correlaciÃ³n de 0.93, lo que indica una corelaciÃ³n lineal fuerte positiva.

A contunuaciÃ³n se realizarÃ¡ el anÃ¡lisis de correlaciÃ³n paso a paso.

```python
#GrÃ¡fico de dispersiÃ³n:
sns.scatterplot(data=data, x='displacement', y='weight')
plt.grid()
plt.show()
```

Del grÃ¡fico de dispersiÃ³n se observa una tendencia a una relaciÃ³n lineal positiva (cuando una de las varaibles aumenta, la otra tambiÃ©n aumenta).

```python
# Normalidad
fig, axs = plt.subplots(1,2, figsize=(10, 4))

axs[0].hist(x=data.displacement, bins=20, alpha=0.5)
axs[0].set_title('DistribuciÃ³n Kilometraje (displacement)')
axs[0].set_xlabel('displacement')
axs[0].set_ylabel('frecuencia')

axs[1].hist(x=data.weight, bins=20, alpha=0.5)
axs[1].set_title('DistribuciÃ³n Peso (weight)')
axs[1].set_xlabel('weight')
axs[1].set_ylabel('frecuencia')

plt.show()
```

De los histogramas se observa que ninguna de las varaibles presenta una distribuciÃ³n normal, aunque sÃ­ se logra notar una leve tendencia a formar una campana de gauss.

Dado que no se cumple con la coindiciÃ³n de normalidad, se procede a calcular el coeficiente de correlaciÃ³n de Spearman.

```python
# CÃ¡lculo de correlaciÃ³n y significancia con Scipy
r, p = stats.spearmanr(data['displacement'], data['weight'])
print(f"CorrelaciÃ³n Spearman: r={r}, p-value={p}")
```

Los test estadÃ­sticos muestran una correlaciÃ³n lineal alta, con claras evidencias estadÃ­sticas de que la relaciÃ³n observada no se debe al azar ( pvalueâ‰ˆ0 ).

Del anÃ¡lisis anterior podemos notar entonces que la correlaciÃ³n entre el peso del vehÃ­culo y su kilometraje es alta, lo que nos darÃ­a a pensar que una variable podrÃ­a ser la causante de la otra, cuando en la prÃ¡ctica no es asÃ­, dado que un vehÃ­culo de mayor peso no necesariamente tendrÃ¡ mayor kilometraje, ni viceversa.

Por lo anterior es importante entender que la correlaciÃ³n entre variables NO implica necesariamente causalidad.
