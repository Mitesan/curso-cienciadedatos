# Dispersión

```python
import pandas as pd
import matplotlib.pyplot as plt
import searborn as sns

plt.style.use("seaborn-dark") #para que los gráficos se visualicen en el tema de fondo negro.
```

```python
df = pd.read_csv('cars.csv')
df.head()
```

```python
df['price_usd'].std() #desviación estandar
```

```python
#Rango = valor max - valor min
rango = df['price_usd'].max() - df['price_usd'].min()
rango
```

```python
#Calcular los quartiles

Q1 = df['price_usd'].quantile(q=0.25)
Q2 = df['price_usd'].median() #Recordamos la mediana Q2
Q3 = df['price_usd'].quantile(q=0.75)

#Calculemos valor max y min con quartiles
min_valor = df['price_usd'].quantile(q=0)
max_valor = df['price_usd'].quantile(q=1)

print(min_valor, Q1, Q2, Q3, max_valor)
```
