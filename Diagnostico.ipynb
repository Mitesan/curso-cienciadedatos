```python
#Importando la libreria Pandas
import pandas as pd
```

```python
#Creando 2 listas
lista1 = ["Rut","Nombre"]
lista2 = [("103458675-3","94857345-0","123847345-5","78606854-2","18555945-3"),("Marta","Raquel","Carmen","Cecilia","Daniela")]
```

```python
#Transformando a diccionario
diccionario = dict(zip(lista1,lista2))
print (diccionario)
```

```python
#Crear DataFrame a partir del diccionario
df = pd.DataFrame(diccionario)
print(df)
```

```python
#Obtener solo los nombres
df["Nombre"]
```

```python
#Obtener los nombres que empiecen con C
df[df.Nombre.str[0]=='C']
```

```python
#Obtener los Rut con digito verificador igual a 3
df[df.Rut.str[-1]=='3']
```
