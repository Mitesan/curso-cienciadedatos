# Mediana, Media

```python
import pandas as pd
import matplotlib.pyplot as plt

pd.options.display.max_columns = None #Para que se visualicen todas las columnas del dataframe
```

```python
df = pd.read_csv('cars.csv') #data de autos usados
df.head()
```

```python
df['price_usd'].mean() #promedio
```

```python
#Nos interesa analizar los atípicos, es decir cuál es el auto con mayor precio. También se puede determinar cuál es el más barato.

df['price_usd'].median() #mediana
#Con la mediana puedo determinar que hay autos costosos a pesar de ser usados.
#El promedio está lejano a la mediana
```

```python
#Vamos a graficar la columna price_usd
df['price_usd'].plot.hist(bins=20, edgecolor="black") #histograma de 20 segmentaciones
plt.show()
```

```python
#la mayoría de los autos están entre 0 y 10000, pero se observan bastantes casos atípicos
```

```python
import seaborn as sns
```

```python
sns.displot(df, x ='price_usd', hue = 'engine_type', multiple = 'stack')
plt.show()

#hue: agrupa por la categoría/columna que se le indique
#multiple='stack' permite visualizar mejor el gráfico, ya que apila las barras de los datos agrupados.
```

```python
#En el gráfico anterior no se visualizan los autos eléctricos, por lo que se intuye que hay muy pocos autos eléctricos.
#El conteo de datos también es un estadístico descriptivo.
#Cuántos autos hay por cada tipo de máquina, de acuerdo a su motor (engine_type) ?

```

```python
df.groupby('engine_type').count()
```

```python
#Solo hay 10 autos eléctricos.
#La estadística descrptiva no es sólo números sino también visualizaciones.
```

```python
#Busquemos un auto que sea de marca Audi y de clase Q7
Q7_df = df[(df.manufacturer_name=='Audi') & (df.model_name=='Q7')]
Q7_df
```

```python
# Modelo por año
sns.histplot(Q7_df, x='price_usd', hue='year_produced')
plt.show()
```

Observaciones:
La mayoría de los autos Q7 que están en ese catálogo
tienden a tener precios bajos, entre 10000 a 20000.
Los colores claros hacen referencia a los modelos más antiguos del 2006, 2007, 2008 el auto más antiguo va depreciando su valor
